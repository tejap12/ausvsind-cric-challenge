I have collected some public data sets online and would like to use them for my analysis. The data set IndAusDS.csv contains ODI matches played between India and Australia from 1971 to 2017. I would like to use this data set to predict winning team in ODI 2019

#### Step1: Load dataset to a data frame

```python
import pandas as pd
ODI_df=pd.read_csv("/Users/nani/Documents/Tejaswini/StatisticsC1/InferentialStatistics/Datasets/IndAusODI.csv",sep=',',engine='python')
ODI_df.head()
```

```python
ODI_df.info()
```

```python
countries=['India','Australia']
IndAus_df=ODI_df[ODI_df['Team 1'].isin(countries) & ODI_df['Team 2'].isin(countries)]
## Only India and Australia ODI matches filtered from main data set
IndAus_df.head(100)
```

```python
IndAus_df.info()
```

```python
IndAus_df['Mdate']=pd.to_datetime(IndAus_df['Match Date'], format="%b %d, %Y")
IndAus_df.head()
```

```python

def winloss (row):
    if row['Winner'] == 'India' :
        return 'Win'
    elif row['Winner'] == 'Australia' :
        return 'Loss'
    elif row['Winner'] == 'no result' :
        return 'Tie'

IndAus_df['WinLoss']=IndAus_df.apply(lambda row: winloss (row),axis=1)
IndAus_df.info()
IndAus_df.head()
```

```python
IndAus_Win_df=IndAus_df.groupby(['Winner','Host_Country'])
IndAus_Win_df.count()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.countplot(x="Winner", data=IndAus_df)
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.countplot(x="Winner", hue='Host_Country', data=IndAus_df)
plt.show()
```

As we see from above plots, the winning rate for ODI matches for Australia is more when compared with India ( Australia won 146 matches and India won 90 matches)
But we have to consider pitch as well, as the host country for Ind vs Aus ODI 2019 is India, We should be considering winning rate in Home pitch
From the above plot, we see Australia has won 52 matches in India and India won 50 matches.


```python
IndAus_Win_df=IndAus_df.groupby(['Winner',IndAus_df['Mdate'].dt.year])
IndAus_Win_df.count()
```

```python
# Year wise winning rate of both the countries
plt.figure(figsize=(14, 10))
sns.countplot(x="Winner", hue=IndAus_df['Mdate'].dt.year, data=IndAus_df)
plt.show()
```

### Predicting Highest run scorer based on below analysis

We have gathered ODI batting data set containing player info, runs, batting information. Proceeding with the analysis to find highest run scorer.

```python
### Loading data set to data frame
bat_df=pd.read_csv("/Users/nani/Documents/Tejaswini/StatisticsC1/InferentialStatistics/Datasets/cricinfo-statsguru-data/ODIs - Batting1.csv",sep=',',engine='python')
bat_df.head()
bat_df.info()
pd.options.display.float_format = '{:,.0f}'.format
```

```python
cols=['Innings Batted','Not Outs','Runs Scored','Highest Innings Score Num','Balls Faced','Hundreds Scored','Scores Of Fifty Or More','Ducks Scored']
bat_df[cols] = bat_df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
bat_df.info()
bat_df.head(100)
```

```python
float_cols=['Batting Avg','Batting Strike Rate']
bat_df[float_cols] = bat_df[float_cols].apply(pd.to_numeric, errors='coerce', axis=1)
bat_df.info()
```

```python
## filter the data based on countries Australia and India
aus_bat_df=bat_df[bat_df.Country=='Australia']
aus_bat_df.info()
```

```python
ind_bat_df=bat_df[bat_df.Country=='India']
ind_bat_df.info()
ind_bat_df.head()
```

Now that we have filtered the dataset based on country, lets filter out players as well who are not active/retired.

```python
## filter out data set to remove inactive/retired players for each country
ind_bat_df=ind_bat_df[~(ind_bat_df['Career End'] < 2018)]
```

```python
aus_bat_df=aus_bat_df[~(aus_bat_df['Career End'] < 2018)]
aus_bat_df.head()
```

```python
ind_bat_df.sort_values(['Runs Scored','Matches Played'],ascending=False).head()
```

```python
aus_bat_df.sort_values(['Runs Scored','Matches Played'],ascending=False).head()
```

```python
aus_bat_df['RunPercent']=aus_bat_df['Runs Scored']/aus_bat_df['Innings Batted']
aus_bat_df.sort_values(by='RunPercent',ascending=False).head(5)
```

```python
ind_bat_df['RunPercent']=ind_bat_df['Runs Scored']/ind_bat_df['Innings Batted']
ind_bat_df.sort_values(by='RunPercent',ascending=False).head(5)
```
